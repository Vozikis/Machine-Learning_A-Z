### Importing the libraries


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Importing the dataset

```python
dataset = pd.read_csv('Credit_Card_Applications.csv')
X = dataset.iloc[:, :-1].values 
y = dataset.iloc[:, -1].values
#we don't train our model. It's unsupervised. We want to get the data of those who got rejected in their application
```

## Feature Scaling


```python
#we normalize our data. 0-1 values
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0,1))
X = sc.fit_transform(X)
```

## Training the SOM


```python
from minisom import MiniSom
#x and y are the dimensions of the SOM. We don't have that many customers so we choose 10x10
#input_len the number of features we have in our dataset. So we need 14+1 = 15
#sigma = the radius of the different neighboors in the grid. default value is 1
#learning_rate = hyperparameter decides how much the weights are updated dfuring each iteration
# The bigger it is the faster it converges. Default values is 0.5
som = MiniSom(x=10, y=10, input_len= 15, sigma= 1.0, learning_rate = 0.5)
#we initialize our weights with random values close to 0 but not 0
som.random_weights_init(X)
#we train our SOM on X. This will apply the steps 4 to 9 for as many iterations as we want
#In this particual case we try with 100 iterations!
som.train_random(data = X, num_iteration = 100)
```

## Visualizing the results


```python
#the whitter it is the closer to the winning neuron

from pylab import bone, pcolor, colorbar, plot, show
bone() #the window that will contain the map. WE just get a white window without anything in it
#put the different winning modes in the map
#som.distance_map() get all the mean distances of the winning nodes
#will return in one matrix so with .T we want them to get them into the correct order. Transpose of matrix
pcolor(som.distance_map().T)
#add a colorbar. Legend of all colours. Those on the right are the normalized mid.
#white colour is the big MID. We know that the frauds will be the ones far from the general rules. So the whiter
#it is the more fraudulent the value. The whitter the more outlier
colorbar()
#the customers that were fraud and got approval have more core value for us. 
#red circles = didn't get approbale
#green squares = got approval
markers = ['o', 's'] #o is a circle and s is the square
colors = ['r', 'g'] #colors for our markers above r = red, g = green
for i, x in enumerate(X): #loop over all customers and get the winning node. we will use red cricle for no approval and grren square for aproval
    #i, x because X is a matrix
    w = som.winner(x) #get the winning node for the first customer. (row)
    #place the marker on the winning mode based on if he got approval or not
    plot(w[0] + 0.5, #put the marker in the center of the winning mode. Center of each square
         w[1] + 0.5, #w[1] is the y coordinate and w[0] is the x coordinate. We add 0.5 so we get to the center
         markers[y[i]], #we need to know if we put red or green. We read in the y[i] vector if he got approval or not
         #if customer didnt get approval then y[i] = 0 so marker[0] will be red circle. if it got it will be green square
         markeredgecolor = colors[y[i]], #we do exacly the same for our colors
         markerfacecolor = 'None', #we only want to color the edge of the marker instead of the whole marker
         markersize = 10, #size of the marker. otherwise it will be very small
         markeredgewidth = 2) #same for the edge
show()#show in order to show the graph

#high risk of fraud for the 1.0 customers
```

## Finding the frauds


```python
mappings = som.win_map(X) #contain's all the different mappings from the winning nodes to the customers. returns a dictionary
#we add only the data the self organized map was used to train our SOM
#we will use concatenate so we have a whole list of the potentional cheaters
#mappings[(k,z)] get the coordinates of the outlier winning node 1,1 for the first one and 4,1 for the second's coordinates
#so we acctually read from the map the winning nodes coordinates with super high value (1) and we jsut type the
#coordinates that we read from the map here in order to get all the customers corresponding to each one of the 
#winning nodes specified!!
frauds = np.concatenate((mappings[(2,5)], mappings[(6,6)]), axis = 0)
#inverse so we get the legit values 
frauds = sc.inverse_transform(frauds)
```

## Printing the Fraunch Clients

```python
print('Fraud Customer IDs')
for i in frauds[:, 0]:
  print(int(i))
```
